```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')
```

```python
#kiểm tra dữ liệu trước khi thực sự đẩy vào Database
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# to_sql dùng để ghi dữ liệu từ dataframe vào database, nếu bảng đã tồn tại thì sẽ bị ghi đè
# head(n=0) makes sure we only create the table, we don't add any data yet.
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

INGESTING DATA IN CHUNKS

```python
# df_iter là một iterator, mỗi lần gọi sẽ trả về một chunk của dataframe với kích thước chunksize đã định nghĩa
# Bản chất nó trả về 1 TextFileReader object, có thể được sử dụng để đọc dữ liệu theo từng phần (chunk) thay vì đọc toàn bộ dữ liệu vào bộ nhớ cùng một lúc.
# Cung cấp dtype để tránh việc phải đọc hết toàn bộ dữ liệu mới xác định được type dữ liệu
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

```python
count = 0
for df_chunk in df_iter:
    count += 1
print(count)    
```

```python
# Lệnh chạy hoàn chỉnh
first = True

for df_chunk in df_iter:

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```

```python
type(df_iter)
```

```python
# cách 2:
# first_chunk = next(df_iter)
first_chunk = next(df_iter) # lấy chunk đầu tiên từ iterator
first_chunk.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace"
)

print("Table created")

first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append"
)

print("Inserted first chunk:", len(first_chunk))

for df_chunk in df_iter: # tiếp tục lấy các chunk còn lại từ iterator, các khối còn lại "biết" tránh khối đầu tiên là nhờ vào cơ chế Con trỏ (Pointer) của đối tượng Iterator (df_iter).
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )
    print("Inserted chunk:", len(df_chunk))
first_chunk = next(df_iter)

first_chunk.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace"
)

print("Table created")

first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append"
)

print("Inserted first chunk:", len(first_chunk))

for df_chunk in df_iter:
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )
    print("Inserted chunk:", len(df_chunk))
```

```python
#Khi bạn xử lý dữ liệu lớn (như file taxi hàng triệu dòng), chương trình thường chạy im lìm rất lâu. tqdm giúp bạn biết được tiến trình đã chạy được bao nhiêu phần trăm, còn bao nhiêu phần trăm nữa thì hoàn thành, và ước tính thời gian còn lại để hoàn thành. Điều này rất hữu ích để bạn có thể theo dõi quá trình xử lý dữ liệu và biết được khi nào sẽ xong.
!uv add tqdm
```

```python
from tqdm.auto import tqdm
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)

# Lệnh chạy hoàn chỉnh
first = True

for df_chunk in tqdm(df_iter):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
